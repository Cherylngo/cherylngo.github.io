# Basic Graph Assignment

```python
#importing matplotlib and pandas
import matplotlib.pyplot as plt
import pandas as pd

#allowing graphs in the program
%matplotlib inline

#importing data as data frames
df_opioid=pd.read_csv('OpidsVA.csv')
df_pres=pd.read_csv('presidents.csv')
df_nsa=pd.read_csv('TOTALNSA.csv')
```

## Opiod Crisis in Virginia

```python
fig,ax=plt.subplots()
ax.scatter(df_opioid['MedianHouseholdIncome'],df_opioid['FPOO-Rate'],color='k',alpha=.4)
fig.suptitle('Opioid Overdoses vs Household Income In Virginia Counties')
fig.set_size_inches(7,5)
ax.set_xlabel('Median Household Income ($)')
ax.set_ylabel('Overdoses (Per 100,000 Residents)')
plt.show()

# A scatterplot is useful for showing the negative relationship between income and opioid overdoses
```

```python
fig,ax=plt.subplots()
ax.scatter(df_opioid['FPOO-Rate'],df_opioid['FFHO-Rate'],color='k',alpha=.45)
fig.suptitle('Heroin Overdoses vs Opioid Overdoses in Virginia Counties')
fig.set_size_inches(7,5)
ax.set_xlabel('Opioid Overdoses')
ax.set_ylabel('Heroin Overdoses')
plt.show()

# A scatterplot is useful for showing the more positive relationship between opioid and heroin overdoses
```

## Presidents and Their Birthplaces

```python
freq=dict(df_pres['State'].value_counts())
fig,ax=plt.subplots()
state=list(freq.keys())
count=list(freq.values())
ax.bar(x = state[:8], height = count[:8], color='black')
fig.set_size_inches(8,5)
fig.suptitle('States With The Highest Number of Presidential Births')
ax.set_xlabel('State')
ax.set_ylabel('Number of Presidents Born')
plt.show()

#A bar plot helps display categorical frequencies such as the number of presidential births in a state
```

## U.S. Vehicle Sales

```python
fig,ax = plt.subplots()
ax.plot(df_nsa['DATE'],df_nsa['TOTALNSA'],color='k',linewidth=.5)
fig.suptitle('U.S. Vehicle Sales Over Time')
fig.set_size_inches(12,6)
ax.set_ylabel('Total Sales ($)')
ax.set_xlabel('Date')
plt.show()

# A line graph displays changes over time in vehicle sales
```
